### Encuesta permanente de hogares (Indec)
#### Módulo de Uso de Tecnologías
##### Objetivo: Breve análisis sobre el acceso y uso de la tecnología e internet en el territorio argentino

```python
import pandas as pd

datai = pd.read_table("EPH_usu_indiv_tic_t421.txt", sep=";")

datai
```

```python
datai.info()
```

```python
usointernet = datai.groupby(datai.IP_III_04).count()
usointernet
```

```python
import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots(figsize=(6,6))
fig1.suptitle("En los últimos meses,¿utilizó internet?")
ax1.pie(usointernet.CODUSU, shadow=True, autopct="%1.0f%%", labels=["Si","No","Ns/Nr"], wedgeprops={"linewidth":2, "edgecolor":"gray"},
        textprops={"fontsize":13})

plt.show()
```

```python
usopc = datai.groupby(datai.IP_III_05).count()
usopc
```

```python
fig2, ax2 = plt.subplots(figsize=(6,6))
fig2.suptitle("En los últimos tres meses, excluyendo el uso de internet,¿utilizó computadora?")
ax2.pie(usopc.CODUSU, autopct="%1.0f%%", labels=["Si","No","Ns/Nr"], wedgeprops={"linewidth":2, "edgecolor":"gray"},
        textprops={"fontsize":13})

plt.show()
```

```python
movil = datai.groupby(datai.IP_III_06).count()
movil
```

```python
fig3, ax3 = plt.subplots(figsize=(6,6))
fig3.suptitle("En los últimos tres meses,¿utilizó un teléfono móvil (celular)?")
ax3.pie(movil.CODUSU, autopct="%1.0f%%", labels=["Si","No","Ns/Nr"], wedgeprops={"linewidth":2, "edgecolor":"gray"},
        textprops={"fontsize":13})

plt.show()
```
